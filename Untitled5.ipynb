```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow  # To display images in Colab
import pandas as pd
import matplotlib.pyplot as plt

# Load the model
model = YOLO('yolov5su.pt')  # You can replace "yolov5su.pt" with the model you want to use

# Video path
video_path = '/content/WhatsApp Video 2024-11-17 at 19.59.49 (1) (1).mp4'
cap = cv2.VideoCapture(video_path)

# Ensure the video is loaded properly
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties like width, height, and frames per second (fps)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Create a list to store the number of people in each frame
person_count_per_frame = []
frame_count = 0

# Path for the output video
output_video_path = '/content/modified_video.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Set the video codec
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Read the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Stop reading when video ends

    # Apply YOLOv5 model on the frame
    results = model(frame)  # Detect people in the frame

    # Get the number of people detected in the frame
    person_count = sum([1 for det in results[0].boxes if det.cls == 0])  # People have class ID = 0

    # Append the count to the list
    person_count_per_frame.append(person_count)

    # Draw the detections on the frame (draw boxes around people)
    annotated_frame = results[0].plot()

    # Write the annotated frame to the new video file
    out.write(annotated_frame)

    # Increment the frame counter
    frame_count += 1

# Release the video objects
cap.release()
out.release()

# Convert the data into a Pandas DataFrame
time_in_seconds = [frame / fps for frame in range(len(person_count_per_frame))]  # Calculate time for each frame
data = pd.DataFrame({
    'Second': time_in_seconds,
    'Person Count': person_count_per_frame
})

# Plot the graph using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(data['Second'], data['Person Count'], marker='o', color='b', linestyle='-', label="Number of People")
plt.title("Number of People Per Second")
plt.xlabel("Time (Seconds)")
plt.ylabel("Number of People")
plt.grid(True)
plt.legend()
plt.show()

# Print the path of the saved video
print(f"The modified video has been saved at {output_video_path}")
```
